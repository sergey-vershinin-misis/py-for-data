Норма - какая-то функция, которая отображает вектор в число

$||.||: V \to R$

Свойства:

$\forall x \in V, \ \forall \lambda \in R: ||\lambda x|| =  |\lambda| \ ||x||$

$\forall x \in V ||x|| > 0 \ и \ ||x|| = 0 \Leftrightarrow x = 0$

$\forall x, y \in V: ||x+y|| < ||x|| + ||y||$

Примеры норм

$||x||_1 = \sum_{i=1}^{n} |x_i|  \ \ \ \ \ || \ [1,2,-3] \ ||_1 = |1| + |2| + |-3| = 1 + 2 + 3 = 6$

$||x||_2 = \sqrt{\sum_{i=1}^{n} x_i^2}  \ \ \ \ \ || \ [1,2,-3] \ ||_2 = 1^2 + 2^2 + (-3)^2 = 1 + 4 + 9 = 14$

$||x||_p = \sqrt[\leftroot{10} \uproot{5} p]{\sum_{i=1}^{n} |x_i|^p}$

$||x||_{\infty} = max_{i} |x_i|$

Можно, имея скалярное произведение, породить норму в пространстве:

$||x|| := \sqrt{\langle x, x \rangle}$

$\sqrt{\langle x, x \rangle} = \sqrt{x_1^2+...+x_n^2} = ||x||_2$

Но не всякая норма порождает скалярное произведение. Например, Манхетонская норма. Тут работает только в одну сторону

Расстояние между двумя векторами
$d(x,y) = ||x-y|| = \sqrt{\langle x-y, x-y \rangle}$

Например, Евклидова норма:

$d(x,y) = ||x-y||_2 = \sqrt{\langle x-y, x-y \rangle} = \sqrt{(x_1-y_1)^2 + ... + (x_n - y_n)^2}$

Скалярное произведение -> Норма -> Расстояние

Гиперплоскость

<img src="./linal/dc_07_3.png" alt="Alternative text" />

$w_1x_1 + ... w_nx_n + b = 0$, где хотя бы один $w_i \neq 0$

w - нормальный вектор к плоскости

Если b не будет, то плоскость всегда будет проходить через ноль.

b - скаляр. Позволяет сдвигать плоскость вдоль вектора нормали. Если мы будем двигать плоскость вдоль плоскости, то ничего не изменится. Так как плоскость бесконечна

$\langle w, x\rangle + b = 0 \ \ \ \ \ w = (w_1,...,w_n)$

Линейная зависимость и независимость

нулевой вектор может быть представлен, как тривиальная комбинация от векторов $x_1, x_2.., x_n$:

$\sum_{i=1}^{n} 0 * x_i = 0$

А что если есть нетривиальная линейная комбинация, котороая приводит к нулю?
Тогда получается линейная зависимость:

$\sum_{i=1}^{n} \lambda_ix_i = 0$, где хотя бы один $\lambda_i \neq 0$. 
Тогда $x_1, x_2.., x_n$ - линейно зависимые. Один вектор может выражаться через другие. 

Это плохо тем, что если мы рассматриваем матрицу признаков, то имея 2 признака, которые линейно зависимы, мы не можем получить единственное решение.

Например, пусть есть рост человека в в одной величине $x_1$ и во второй $x_2$. причем $x_2 = 2x_1$

Тогда напишем линейную зависимость

$w_1x_1 + w_2x_2 = y$

Пусть $w_1  = 0.3$, а $w_2 = 0.35$.  тогда

$0.3x_1 + 0.35x_2 = y \rightarrow 0.3x_1 + 0.35*2*x_1 = y \rightarrow 0.3x_1 + 0.7x_1 = y$

Но к такому же значению может привести и другая комбинация

$0.2x_1 + 0.4x_2 = y \rightarrow 0.2x_1 + 0.4*2*x_1 = y \rightarrow 0.2x_1 + 0.8x_1 = y$

Поэтому решение не будет единственным. Это плохо

Если существует только тривиальное решение для того, чтобы получить ноль в линейной комбинации, то векторы линейно независимые

Отличие от пораждающего множества в том, что пораждающее множество может быть больше N.

Детерминант - численный способ охарактеризовать линейную трансформацию
1. абсолютное значение - какая площадь поменялась после преобразования
2. знак - в какую сторону преобразование сработало

$
det(
\begin{bmatrix}
    a & b \\
    c & d
\end{bmatrix}
) =|ad - bc|
$

$
det(
\begin{bmatrix}
    a & b & c \\
    d & e & f \\
    g & h & j \\
\end{bmatrix}
) = a(ej-fh) -b(dj-fg) + c(dh-eg)
$

Не имеет значение по строке или колонке идет расскрытие.

Детерминант не меняется, если производить базовые операции над строками или столбцами. Иногда удобно преобразовать матрицу к треугольному виду. Меньше расчетов.

Также детерминант не меняется при транспонировании.

$det A = det A^T$

$det AB = det A * det B$

$det A^{-1} = \frac{1}{det A}$
